## Feature Engineering
### Adding columns for "Total Space" And "Average"

```python
df["Total Score"] = df["math score"] + df["reading score"] + df["writing score"]
df["Average"] = df["Total Score"] / 3
df.head(2) 
```

```python
reading_full = df[df["reading score"] == 100]["gender"].count()
writing_full = df[df["writing score"] == 100]["Average"].count() ## gender and Average just for count()
math_full = df[df["math score"] == 100]["Average"].count()

reading_full, writing_full, math_full
```

```python
reading_less = df[df["reading score"] <= 20]["lunch"].count()
writing_less = df[df["writing score"] <= 20]["lunch"].count()
math_less = df[df["math score"] <= 20]["lunch"].count()

reading_less, writing_less, math_less
```

```python
fig, axs = plt.subplots(1, 2, figsize = (10, 4))
plt.subplot(121) # plt.subplot(rows, columns, position)
sns.histplot(data = df, x= "Average", bins = 30, kde = True, color = "r")
plt.subplot(122) # plt.subplot(rows, columns, position)
sns.histplot(data = df, x = "Average", kde = True, color = "g", hue = "gender")
plt.show()
```

```python
fig, axs = plt.subplots(1, 2, figsize = (10, 4))
plt.subplot(121) # plt.subplot(rows, columns, position)
sns.histplot(data = df, x= "Total Score", bins = 30, kde = True, color = "r")
plt.subplot(122) # plt.subplot(rows, columns, position)
sns.histplot(data = df, x = "Total Score", kde = True, color = "g", hue = "gender")
plt.show()
```
